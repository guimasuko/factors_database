## By Permno

O objetivo desse notebook é criar um conjunto de tabelas na extensão .csv onde cada tabela representa uma série diária de uma ação a partir de dois dataframes, um contendo dados diários de firmas e suas caracaterísticas e outro contendo dados de localização em percentis de fatores financeiros.

E.g. '10001.csv' será uma planilha (tabela) contendo a série temporal da ação cujo permno é o 10001.

```python
crsp = pd.read_csv('../input/table.csv')
```

```python
factors = pd.read_csv('../input/table2.csv')
```

```python
factors
```

```python
# todos permnos no dataframe factors
factors_permno = factors['permno'].unique()
```

```python
# todos permnos no dataframe crsp
crsp_permno = crsp['PERMNO'].unique()
```

```python
# somente os permnos que estão em ambos os dataframes
permnos = np.intersect1d(factors_permno, crsp_permno)
```

```python
permnos
```

```python
for i in permnos:
    """
    Primeiro, criaremos um dataframe com os dados CRSP para cada PERMNO
    Neste dataframe, indexaremos o DIA (nessa base os dados já se encontram na frequência diária)
    """
    df = crsp[crsp['PERMNO'] == i]
    df['date'] = df['date'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m%d'))
    df.set_index('date', inplace=True)

    """
    Segundo, criaremos um dataframe com os dados dos FACTORS para cada PERMNO
    Neste dataframe, indexaremos o DIA (nessa base os dados se encontram na frequência mensal)
    """
    df_ = factors[factors['permno'] == i]
    df_['date'] = pd.to_datetime(df_['date'])
    df_.set_index('date', inplace=True)

    start = min(df_.index[0], df.index[0]) # pegando a primeira data entre as primeiras datas dos dois dataframes
    end = max(df_.index[len(df_) - 1], df.index[len(df) - 1]) # pegando a última data entre as últimas datas dos dois dataframes
    days = pd.date_range(start = start, end = end, freq='D')
    
    df_ = df_.reindex(days, method='ffill') # função que reindexa o dataframe para frequência diária utilizando o método de preenchimento para as novas linhas

    """
    Por fim, agora temos dois dataframes na mesma frequência.
    Um com dados CRSP e outro com dados dos FACTORS
    Basta dar um merge entre os dois e finalizamos a iteração fazendo o download em csv do dataframe
    """
    df_final = pd.merge(df, df_, left_index = True, right_index = True)
    df_final = df_final.drop('permno', axis=1)
    output_path = "..\\output\\by_permno\\{}.csv".format(i)
    df_final.to_csv(output_path, sep=',', encoding='utf-8')
```
